```python
import numpy as np
import matplotlib.pyplot as plt
tr_acc = np.load('tr_acc_new.npy')*2
va_acc=np.load('va_acc_new.npy')*2
tr_loss = np.load('tr_loss_new.npy')*2
va_loss =np.load('va_loss_new.npy')*2

print(np.max(va_acc),np.max(tr_acc))
plt.figure()
plt.plot(tr_acc[0])
plt.plot(va_acc[0])
plt.plot(tr_acc[1])
plt.plot(va_acc[1])
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(['train1', 'validation1', 'train2', 'validation2'])
plt.savefig('./acc_with_ray.png')
plt.show()

plt.plot(tr_loss[0])
plt.plot(va_loss[0])
plt.plot(tr_loss[1])
plt.plot(va_loss[1])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(['train1', 'validation1', 'train2', 'validation2'])
plt.savefig('./loss_with_ray.png')
plt.show()
```
